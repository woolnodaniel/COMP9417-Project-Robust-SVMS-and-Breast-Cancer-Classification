```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
ds = pd.read_csv('data.csv')
ds.info()
```

```python
ds
```

```python
# Notice the "Unnamed" column: contains only NaN and no data. We drop it
# Also don't need to keep the ID column: does not affect the diagnosis
ds = ds.drop(['Unnamed: 32', 'id'], axis=1)
```

```python
# transform any non-numeric values into numeric values
# In this case, only applies to the "diagnosis" column
# Classes distinguised as +/- 1, for use later with SVM implementations.
ds.diagnosis = ds.diagnosis.map(lambda x: 1 if x == 'M' else -1)
```

```python
ds
```

```python
# check for NaN values
ds.isnull().values.any()
```

```python
# normalise the data: each column is recentred at 0 with standard deviation, but NOT the diagnosis
for col in ds.columns:
    if col == 'diagnosis':
        continue
    mu = np.mean(ds[col])
    sigma = np.std(ds[col])
    ds[col] = ds[col].map(lambda x: (x-mu)/sigma)
ds
```

```python
# feature selection: which features to use?
corr_col = ds.corr()['diagnosis']
corr_col
plt.figure()
heat_map = sns.heatmap([corr_col], square=True, cbar_kws=dict(use_gridspec=False, location="top"))
heat_map.set_yticklabels(heat_map.get_yticklabels(), rotation=0)
plt.show()
```

```python
# can see that columns 10, 12, 15, 19, 20 are not correlated at all with the result... drop columns
# this can be done formally by setting a lower bound on what qualifies as "correlated"; in this
#    case, if the correlation is less than 0.1. 
drop_index = []
for i in range(len(corr_col)):
    if corr_col[i] < 0.1: #consider the feature uncorrelated with target
        drop_index.append(i)
print(drop_index)
```

```python
drop_list = [ds.columns[i] for i in drop_index]
ds = ds.drop(drop_list,axis=1)
ds
```

```python
# do we need to worry about class imbalance?
ds.diagnosis.value_counts()
```

```python
# not imbalanced! (63% malignant, 37% benign). No resampling needed. 
```

```python
# save as new datafile for manipulation across different languages (save more preprocessing):
file = open("processed_data.csv", "w")
ds.to_csv(file, index=False)
file.close()
```
